### Borracho en 2 Dimensiones.

```python
#para dummies
n = 500
x,y = 0,0
pasosx = []
pasosy = []

for i in range(n):
    x += 2 * np.random.random() - 1
    y += 2 * np.random.random() - 1
    if x < 0:
        x=0
    if y < 0:
        y=0
    if x > 5:
        x=5
    if y > 5:
        y=5
    pasosx.append(x)
    pasosy.append(y)
    
    
    
plt.plot(pasosx, pasosy, '#00cc66')
    
```

```python
n = 65
for j in range(10):
    x , y = np.random.randint(5) , np.random.randint(5)
    pasosx = []
    pasosy = []
    
    for i in range(n):
        x += 2 * np.random.random() - 1
        y += 2 * np.random.random() - 1
        if x < 0:
            x=0
        if y < 0:
            y=0
        if x > 5:
            x=5
        if y > 5:
            y=5
        pasosx.append(x)
        pasosy.append(y)



    plt.plot(pasosx, pasosy, colors[j])
```

### Borracito Arreglado

```python
n = 65
for j in range(10):
    x , y = np.random.randint(5) , np.random.randint(5)
    pasosx = np.zeros(n)
    pasosy = np.zeros(n)
    
    for i in range(n):
        x += 2 * np.random.random() - 1
        y += 2 * np.random.random() - 1
        if x < 0:
            x=0
        if y < 0:
            y=0
        if x > 5:
            x=5
        if y > 5:
            y=5
        pasosx[i] = x
        pasosy[i] = y



    plt.plot(pasosx, pasosy, colors[j])
```

```python
n = 65
for j in range(10):
    x , y = np.random.randint(5) , np.random.randint(5)
    pasos = np.zeros([n, 2])
    
    for i in range(n):
        x += 2 * np.random.random() - 1
        y += 2 * np.random.random() - 1
        if x < 0:
            x=0
        if y < 0:
            y=0
        if x > 5:
            x=5
        if y > 5:
            y=5
            
        pasos[i][0] = x
        pasos[i][1] = y



    plt.plot(pasos[:,0], pasos[:,1], colors[j])
```
